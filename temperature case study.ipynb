```python
# Temperature Monitoring and Control System

# Function to classify temperature
def classify_temperature(temp):
    if temp < 18:
        return "Cold"
    elif 18 <= temp <= 28:
        return "Normal"
    else:
        return "Hot"


# Function to calculate statistics
def calculate_statistics(readings):
    average = sum(readings) / len(readings)
    highest = max(readings)
    lowest = min(readings)
    return average, highest, lowest


# Function to display report
def display_report(readings, counts):
    average, highest, lowest = calculate_statistics(readings)

    print("\nTemperature Summary")
    print("-------------------")
    print(f"Average Temperature = {round(average, 2)}")
    print(f"Highest Temperature = {highest}")
    print(f"Lowest Temperature = {lowest}")

    print("\nCondition Count")
    print(f"Hot = {counts['Hot']}")
    print(f"Normal = {counts['Normal']}")
    print(f"Cold = {counts['Cold']}")

    print("\nSuggested Action")

    action_taken = False

    if counts["Hot"] > 0:
        print("Fan ON")
        action_taken = True

    if counts["Cold"] > 0:
        print("Heater ON")
        action_taken = True

    if not action_taken:
        print("No action required")


# Main Program
def main():
    readings = []
    counts = {"Hot": 0, "Normal": 0, "Cold": 0}

    num = int(input("Enter number of readings: "))

    for i in range(num):
        temp = float(input("Enter temperature: "))
        readings.append(temp)

        condition = classify_temperature(temp)
        counts[condition] += 1

    display_report(readings, counts)


# Run program
if __name__ == "__main__":
    main()
```
